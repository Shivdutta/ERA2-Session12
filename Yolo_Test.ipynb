```python
%matplotlib inline
import matplotlib.pyplot as plt
import os
from PIL import Image
from IPython.display import display, HTML

# Define the CSS style to remove scrollbars
remove_scrollbars_css = """
<style>
.output {
    overflow: visible !important;
}
</style>
"""

display(HTML(remove_scrollbars_css))

image_dir = "/content/drive/MyDrive/Colab_Notebooks/Session12/YoloV3-master/test_out"

image_files = os.listdir(image_dir)

image_files = image_files[:16]

fig, axes = plt.subplots(4, 4, figsize=(20, 20))

for i, image_file in enumerate(image_files):
    # Load image
    image_path = os.path.join(image_dir, image_file)
    image = Image.open(image_path)

    # Plot image
    row = i // 4
    col = i % 4
    axes[row, col].imshow(image)
    axes[row, col].axis('off')

plt.tight_layout()
plt.show()
```
